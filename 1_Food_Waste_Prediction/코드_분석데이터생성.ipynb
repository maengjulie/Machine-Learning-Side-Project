## 카드 데이터 합치기

```python
df3 = pd.read_csv('/content/drive/My Drive/Colab Notebooks/04_음식관련 카드소비_CARD_SPENDING.CSV',encoding='euc-kr')
df3
```

```python
df2_1.head()
```

```python
df3 = df3[df3['emd_nm']!='추자면']
df3
```

```python
df3 = df3[df3['emd_nm']!='우도면']
df3
```

```python
df3 = df3.drop(['emd_cd'], axis=1)
df3
```

### 외식(업종코드 참고해서 분류) : 한식+패스트푸드+간식+아시아음식+양식+주점 및 주류판매 + 뷔페 

```python
eat_out = df3[(df3['mct_cat_cd'] == 'C00100') | (df3['mct_cat_cd'] == 'C00500') | (df3['mct_cat_cd'] == 'C00400') | (df3['mct_cat_cd'] == 'C00300') | (df3['mct_cat_cd'] == 'C00200') | (df3['mct_cat_cd'] == 'C00900') | (df3['mct_cat_cd'] == 'C00400') | (df3['mct_cat_cd'] == 'C00300') | (df3['mct_cat_cd'] == 'C00200') | (df3['mct_cat_cd'] == 'C00700')]
eat_out
```

```python
make_group = eat_out[['use_cnt','use_amt']].groupby([eat_out['base_date'],eat_out['emd_nm']])
eat_out2 = pd.DataFrame(make_group.sum()).reset_index()
eat_out2
```

### 장보기: 마트/슈퍼마켓 + 식품 + 농축수산물

```python
market = df3[(df3['mct_cat_cd'] == 'A00200') | (df3['mct_cat_cd'] == 'C01000') | (df3['mct_cat_cd'] == 'C01200')]
market
```

```python
make_group = market[['use_cnt','use_amt']].groupby([market['base_date'],market['emd_nm']])
market2 = pd.DataFrame(make_group.sum()).reset_index()
market2
```
