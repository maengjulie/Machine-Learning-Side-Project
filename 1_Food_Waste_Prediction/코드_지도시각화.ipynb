```python
import plotly
import plotly.express as px
import json
import mapboxgl
import os
import plotly.graph_objects as go
import folium
```

```python
data = pd.read_csv('G2.csv',encoding='euc-kr')
data.head()
```

```python
data = data[(data['em_g'] != 0) & (data['emd_nm'] != '알수없음')]
data['emd_nm']= data['emd_nm'].replace(['일도1동','일도2동', '이도1동', '이도2동', '삼도1동', '삼도2동', '용담1동', '용담2동'], ['일도일동', '일도이동', '이도일동', '이도이동', '삼도일동', '삼도이동', '용담일동', '용담이동'])
```

```python
data[data['emd_nm'] == '일도1동']
```

```python
data.to_csv('emg.csv',encoding='utf-8')
```

```python
data1 = data.drop(['Unnamed: 0','base_year','base_month','base_date','resid_reg_pop','foreign_pop','total_pop','외식_use_amt','장보기_use_amt','배달_use_amt'], axis=1)
data1
```

```python
geo_path2 = 'LSMD_ADM_SECT_UMD_50 (3).zip.geojson'
geo_data2 = json.load(open(geo_path2, encoding='UTF-8'))
geo_data2
```

```python
import plotly.graph_objects as go
```

```python
import requests
import json
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
import os
import webbrowser
import folium
from folium import plugins
```

```python
m = folium.Map(location=[33.4, 126.5], tiles="OpenStreetMap", zoom_start=7)

m.choropleth(
    geo_data=geo_data2,
    name='배출량',
    data=data1,
    columns=['emd_nm', 'em_g'],
    key_on='feature.properties.EMD_NM',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.3,
    legend_name = '배출량'
)

folium.LayerControl().add_to(m)

# Save to html
m.save('folium_kr.html')
webbrowser.open_new("folium_kr.html")
```
